```python
#모든 텍스트가 contents이라는 하나의 긴 문자열로 들어가서, 
# contents[n] 으로 한글자씩 불러서 사용
#라인단위의 데이터에서는 오른쪽끝 공백이 포함된상태로 문자열이 생성
filename = 'exam.txt'
with open(filename) as file_object:
    contents = file_object.read()

print(contents)
```

```python
with open('exam.txt') as data:
    contents = data.read()

print(contents)
```

```python
data = open('exam.txt', 'r')
contents = data.read()
data.close()

print(contents)
```

```python
# 라인 단위로 리스트를 생성
with open('exam.txt') as data:
   lines = data.readlines()

print(lines)
# n번째 줄이 lines[n-1]에 문자열로 들어감. 
# 문자열을 읽는 인덱스까지 포함하면 lines[n][m] 처럼 
# 2차원 리스트나 배열처럼 불러서 사용할수 있음.
# 오른쪽끝에 공백문자는 포함된 상태임
```

```python
with open("exam.txt") as f:
    lines = f.readlines()
    string = ''.join([line.rstrip() for line in lines])
print(string)

# 여러라인을 하나로 합칠때는 라인끝마다 공백이 있는데,  
# .rstrip()이라는 함수로 제거해주어야 함
# .split()을 했을경우는 rstrip()을 따로 해주지 않아도 자동으로 제거
```

CSV 파일 읽기/쓰기

```python
import csv
f = open('sales.csv', 'r', encoding='utf-8')
rdf = csv.reader(f)
#for line in rdf:
#  print(line)
count=0
for i in rdf: 
    print(i)
    count+=1
    if(count==5): break
# 한줄한줄 리스트 형태로 읽음
f.close()
```

```python
with open('sales.csv') as csvfile:
    rdr = csv.DictReader(csvfile)  
#DictReader : 딕셔너리 데이터 형태로 읽음.
    count=0
    for i in rdr:
        print (i)
        count+=1
        if(count==5): 
            break
```

```python
import csv
f = open('text_write.csv', 'w', encoding='utf-8') #  파일생성하고 열어줌
wr = csv.writer(f)
wr.writerow([1,"Alice",True])
wr.writerow([2,'Bob',False])
# wirterow함수에 줄바꿈이 포함되어
f.close()
```
