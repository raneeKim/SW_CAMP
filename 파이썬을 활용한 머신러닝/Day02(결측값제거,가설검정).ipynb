### 결측값 제거
    - 결측값 제거 : 결측값이 존재하는 행을 제거
    - 결측값 대치 : 결측값을 다른 값으로 채워 넣음
        - 단순대치 : 평균/중앙값/최빈값과 같은 데이터의 대표값으로 대치
        - 알고리즘을 이용한 대치 : 학습을 통해 주변값과 유사한 다른 값으로 대치
        - 보간법 : 특정 간격을 가진 결측값에 대해 순차적으로 결측값을 보간(대치) / 시간,날짜,주기

```python
df1 = pd.read_csv('data\\01_Data.csv')
print(df1.shape)
df1.head(2)
```

```python
df1.isnull().sum()
```

```python
# 결측값의 비율이 많은 항목을 제외하고 분석 (그 항목이 분석에 있어 중요하지 않다면!!!!)
df1.drop(columns=['Age', 'Credit_Rank','Bank']).dropna()
```

```python

df1['Credit_Rank'].tail(10)
```

```python
# 결측값 대치
df1['Credit_Rank'].tail(10).fillna(0)
```

```python
# 결측값 대치 (평균 대치)
df1['Credit_Rank'].tail(10).fillna(df1['Credit_Rank'].mean())
```

```python
# 결측값 대치
df1['Credit_Rank'].tail(10).fillna(method='bfill') # 아래 있는 실제값을 기준으로 위로 다음 실제값이 등장할 때 까지 대치
```

```python
# 결측값 대치
df1['Credit_Rank'].tail(10).fillna(method='ffill') # 위에 있는 실제값을 기준으로 다음 실제값이 등장할 때 까지 대치
```

### CDA

- 통계적 가설 검정 : 규명하고자 하는 바를 **가설**로 수립하여, 해당 가설이 참인지 거짓인지 **객관적인 값**으로 규명하는 작업
- 가설 :
  - 귀무가설 : 현재까지 주장되어 온 것이거나, 현재까지 유지되었던 상황에 변화나 차이가 없는 상태를 설명하는 가설
        - 현재까지 유지된 것 -> 보통상황 (각 항목들이 독립된 상황)
  - 대립가설 : 새로 주장하는 것, 표본으로 부터 현자와 다른 상황 변화가 발생한 상태
 
- 10대와 20대의 연령대 간 주당 방문횟수의 차이가 있는지 검정
  - 귀무가설 : 10대와 20대 간의 주당 방문 횟수의 평군의 차이가 없다.
  - 대립가설 : 10대와 20대 간의 주당 방문 횟수의 평균의 차이가 있다.
