```python
p7 = df_process[['datetime','A1','A2','A3','A4','A5','A6']].set_index('datetime')
p7
```

```python
# 한시간 단위로 데이터 상에 존재하는 연속형 자료들의 중앙값을 계산
p8 = p7.resample('1H').median().reset_index()
```
