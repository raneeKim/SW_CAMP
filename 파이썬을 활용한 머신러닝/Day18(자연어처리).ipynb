# 자연어 처리
- 자연어(Natural Language) : 사람들이 일상적으로 사용하는 언어
- 자연어 처리(Natural Language Preprocessing, NLP) : 컴퓨터가 사람의 언어를 이해하거나 생성할 수 있도록 데이터를 처리하는 기술
- 컴퓨터 관점에서 자연어 처리
    - NLU(Natural Language ***Understanding***): 언어 -> 컴퓨터/ 컴퓨터가 사람의 언어를 이해하는 과정
    - NLG(Natural Language ***Generation***): 컴퓨터 -> 언어 / 컴퓨터가 사람의 언어를 생성하는 과정
    - 전산 언어학(Computational Linguistics)
        - 컴퓨팅 기술을 언어학에 적용하여 사용하는 분야
        - 컴퓨터 공학(프로그래밍) + 언어학 + 통계학

- 자연어 처리의 난이도가 높은 이유
    1. 언어의 다양성: 각 언어의 특성이 서로 다름에서 오는 어려움(영어/ 한국어/ 라틴어/ ...)
    2. 언어의 중의성: (ex: 참 잘~ 한다.) -> 긍/부정의 의도 해석
    3. 규칙 예외 -> 표준어/ 유행어/ 줄임말

- 자연어 처리의 Trand
    1. 규칙 기반 : 문법의 규칙을 사전에 정의하여, 규칙을 기반으로 자연어를 처리
        - 언어의 중요성/ 규칙예외
    2. 통계기반 : 확률을 이용해 문장에 오는 단어에 대한 확률/ 빈도 등을 계산하여 처리
        - 대량의 문장분석/ 의도파악
    3. 데이터 마이닝 기반 (텍스트 마이닝) : 전수 조사 중심의 학습 체계를 통한 텍스트 마이닝 기술
        - 자연어 생성(기계 번역)
    4. 신경망 알고리즘 (딥러닝) : 자연어 이해에서 생성까지 (Text to Text) / Image to Text / Text to Image

# 자연어 처리 연산
- 전산언어학에서 사용하는 자연어 처리의 연산
    - Edit Distance : 2개의 단어가 얼마나 다른지를 계산한 척도
        - Cat / Hat: 1
        - 단어의 유사성을 파악할 때 사용
    - Stemming : 어간 추출, 단어의 본래 형태를 추출
        - Computer -> Compute
        - Going -> GO
        - 문장의 의도파악 사용/ 언어 사전이 필요
    - Lemmatization : 문장속에서 다양한 형태로 활용 된 단어의 표제어(Lemma)를 찾는 연산
        - 표제어 : 단어 사전에 등록되어 있는 단어의 기본형
        - 문장의 의도파악 및 단어의 유사성 파악

# Continuous Bag of Words (CBOW)
- 문서 내의 모든 단어를 모아넣고, 단어에 빈도수를 추출하여 표현하는 기법
- 장점 :
  - 문장에 대한 구분 / 문장의 특성을 잘 표현
  - 비교적 넓은 범위(긴 문장)에서 사용이 가능 / 쉽고 빠른 구축 가능
- 단점 :
  - 단어의 문맥적 의미를 반영하기는 어려움
  - 문장이 짧은 경우엔, 문장 자체를 표현하는 것에 한계가 있음 (Sparse Matrix)
    - 대부분의 행렬값이 0값으로 채워져서 분석의 속도와 성능을 저하

```python
from konlpy.tag import Kkma
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text5 = '''너와 함께 한 시간 모두 눈부셨다.
날이 좋아서, 날이 좋지 않아서, 날이 적당해서, 모든 날이 좋았다.
그리고 무슨일이 벌어져도 니 잘못이 아니다'''
```

# Skip-Gram
- 중심 단어로 부터 주변단어들을 **예측** / 주어진 단어로 주변에 어떤 단어들이 올지 확률값으로 계산
- Skip-Gram 단어 표현 기법 종류 중 하나이나, 신경망 알고리즘을 이용해 분석
- 신경망 (입력층 -> 처리층 -> 출력층)
- 장점 :
  - 의미론적 관계 학습 : 단어간의 복잡한 의미를 파악하여 단어간의 패턴이나 관계를 학습
  - 드물게 나타나는 단어를 처리

```python
with open(r'C:\Users\UserK\Desktop\Ranee\data\ML\data3.txt','r',encoding='utf-8') as file :
    text3 = file.read()
text3
```

```python
from gensim.models import Word2Vec # gensim : genrate similar
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent_tokenize(text3)
```

```python
[okt.nouns(x) for x in sent_tokenize(text3)]
```

```python
# 1. 띄어쓰기 또는 온점을 기준으로, 문장을 분할 => sent_tokenize(text3)
# 2. 분리된 문장에서 명사만 추출 => okt.nouns(x)
sent_nouns = [okt.nouns(x) for x in sent_tokenize(text3)]
```

```python
# skip Gram 학습
model = Word2Vec(sent_nouns, vector_size=100, window=5, min_count=1, sg=1)
# vector_size = : 학습된 각 단어의 벡터 차원 수를 설정 (단어의 다양성이 많을 수록 큰 값을 부여)
# window = : 모델 학습 시, 반영되는 컨텍스트 윈도의 크기 (중심단어를 기준으로 좌우의 몇개 단어까지 반영할건지)
# min_count = : 모델이 학습에 포함시킬 단어 최소 빈도 수 (min_count=1 : 데이터 단어가 한번만이라도 등장했다면 학습에 포함)
# sg = : 단어표현을 어떻게 수행할지 결정 (sg=1 : skip gram 방식 활용해 단어를 예측)
```

```python
# 단어 간 유사도 계산
model.wv.similarity('신호','정보')  # '신호'라는 단어와 '정보'라는 단어의 유사도를 계산
# Cosine Similarity (코사인 유사도)로 계산 : -1 ~ +1 
# +1 두 단어가 매우 유사하고 관련이 깊다
# 0 두 단어는 전혀 관련이 없다 (독립이다)
# -1 두 단어가 서로 반대되는 관계에 있다 => 이론상으로는 도출되지만 실제로는 나오지않음
```

```python
# 특정 단어의 유사한 단어들을 모두 확인
model.wv.most_similar('정보')
```
