# Ensemble 

- 앙상블 알고리즘 : 여러개의 모델에 의한 결과를 종합하여, 단일 대비 모델의 성능을 더 높이는 결합 방법
  - 전체 데이터에서 적절한 표본 추출로 여러개의 훈련용 데이터를 만들고, 각 데이터에 각각의 모델을 다르게 적용하여 학습
  - 여러 모델의 예측값을 가중투표(Weighted Vote)를 이용해 결합
  - 이전 모델의 오차를 줄이는 방향으로 파라미터를 조정한 새로운 개선된 모델을 탐색

- 장점 :
  - 다양한 모델의 예측 결과를 결합/선택 함으로써 단일 모델로 학습 했을 때 보다 높은 성능을 보임
  - 전체적인 분산을 감소시킴으로 성능을 높임
  - 이상치에 대한 대응 능력이 높음
  - 과적합에 대한 대응

- 단점 :
  - 시간이 오래 걸림 / 해석이 어려움
 
0. 보팅(Voting) : 여러 다른 모델의 예측을 결합하여 최종 예측을 만드는 기법. Voting은 주로 분류(Classification) 문제에서 사용되며, 각 개별 모델의 예측에 가중치를 부여하거나 단순한 투표로서 최종 예측을 수행
1. 배깅(Bagging): 배깅은 Bootstrap Aggregating의 약자로, 부트스트랩 샘플링(복원 추출 / Sub Sampling)을 통해 여러 개별 모델을 생성하고, 이들의 예측을 평균이나 다수결 투표 등으로 결합하여 최종 예측.
2. 부스팅(Boosting): 약한 학습기(weak learner)를 순차적으로 학습시켜 강력한 모델을 구성. 이전 모델이 잘못 예측한 샘플에 가중치를 더 주어, 다음 모델이 더 잘 예측 ( AdaBoost, Gradient Boosting, XGBoost, LightGBM)

# Random Forest 

- 여러개의 Decision Tree 을 묶어 과적합을 회피할 수 있는 방향으로 설계된 Bagging 계열의 알고리즘
- Bagging + Random Selection ( X를 Random으로 선택하여 학습)
- 원리
  - Bootstrap Sampling
    - Random Forest(RF) 원본데이터에서 무작위로 복원추출(Sub Sampling)을 통해 여러개의 Bootstrap Data를 생성
    - 각 Bootstrap Data 원본 데이터에서 무작위로 (앞서 학습된 데이터와 무관하게) Sampling
  - Random Selection
    - 각 Decision Tree(DT) 학습을 수행 할 때, 무작위로 X를 선택하여 일부 특성들에 대해서만 학습이 수행되도록 Node를 분할
    - 각각의 DT 서로다른 특성을 사용하도록 만들어, 과적합을 방지하고 모델의 다양성을 증대
  - Multiple Decision Tree
    - RF는 다수의 DT를 생성 / 각 트리는 Bootstrap Data를 기반으로 학습
  - Voting & Averaging
    - 분류문제에서는 다수결 투표 (Majority Voting) 통해 각 트리의 예측 결과를 결합하여 최종적으로 예측
    - 회귀문제에서는 각 트레의 예측값의 평균을 계산하여 최종 예측을 수행

```python
# UnicodeDecodeError : 데이터 파일 내 제3기호(숫자와 영어를 제외한 나머지 언어)에 대한 변환식이 설정되어 있지 않는 경우 
df_charge = pd.read_csv('18_Charge_data.csv',encoding='cp949') # cp949 : 한글 데이터에 대한 인코딩 수식 
df_custom = pd.read_csv('18_Customer_data.csv',encoding='cp949')
```

```python
df_custom['고객이탈여부'].unique()
```

```python
df_custom['Target'] = df_custom['고객이탈여부'].replace({'No' : 0 , 'Yes' : 1})
df_custom['Target'].isnull().sum()
```

```python
df_custom.columns
```

```python
Y = df_custom['Target']
X = df_custom[['성별', '연령', '결혼여부', '부양자유무', '부양자수','보안서비스', '백업서비스', '기술지원서비스', '데이터무제한']]
```

```python
from sklearn.model_selection import train_test_split
from imblearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from imblearn.under_sampling import TomekLinks
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=1234)
```

```python
numeric_list = X.describe().columns
category_list= X.describe(include='object').columns
```

```python
numeric_pipe = make_pipeline(SimpleImputer(strategy='median'))
category_pipe= make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder())
prepro_pipe  = make_column_transformer((numeric_pipe, numeric_list), (category_pipe, category_list))
```

```python
model_pipe1 = make_pipeline(prepro_pipe,TomekLinks(), DecisionTreeClassifier())
model_pipe2 = make_pipeline(prepro_pipe,TomekLinks(), RandomForestClassifier())
```

```python
# Decision Tree 
hyperparameter_DT = {'decisiontreeclassifier__max_depth':range(5,10), # depth 낮을수록 과적합을 피할 수 있음 
                    'decisiontreeclassifier__criterion':['gini','entropy'],
                    'decisiontreeclassifier__class_weight':[None, 'balanced']}
grid_model_DT = GridSearchCV(model_pipe1, param_grid=hyperparameter_DT, cv=3, scoring='f1', n_jobs=-1)
grid_model_DT.fit(X_train, Y_train)
best_model_DT = grid_model_DT.best_estimator_
pickle.dump(best_model_DT, open('model_DT.sav','wb'))
```

```python
# Random Forest 
hyperparameter_RF = {'randomforestclassifier__max_depth':range(5,10), # depth 낮을수록 과적합을 피할 수 있음 
                    'randomforestclassifier__criterion':['gini','entropy'],
                    'randomforestclassifier__class_weight':[None, 'balanced']}
grid_model_RF = GridSearchCV(model_pipe2, param_grid=hyperparameter_RF, cv=3, scoring='f1', n_jobs=-1)
grid_model_RF.fit(X_train, Y_train)
best_model_RF = grid_model_RF.best_estimator_
pickle.dump(best_model_RF, open('model_RF.sav','wb'))
```

```python
# 평가
from sklearn.metrics import classification_report
```

```python
def eval_func(model) :
    Y_train_pred = model.predict(X_train)
    Y_test_pred = model.predict(X_test)
    print('학습성능')
    print(classification_report(Y_train,Y_train_pred))
    print('일반화성능')
    print(classification_report(Y_test,Y_test_pred))
```

```python
eval_func(best_model_DT)
```

```python
eval_func(best_model_RF)
```

- **Random Forest Hyperparameter**
    - n_estimators : 의사결정 DT 모델의 개수를 지정 (기본값 100)
    - max_feature  : 각 노드에서 고려할 특성(X)의 최대 개수를 지정 / (기본값 'auto' / 정수 / 'log2', 'sqrt')
    - boostrap : 데이터 샘플을 복원 추출 할지 말지 결정 (기본값 True / False -> 복원추출 X / Random selection)  
    - criterion / max_depth / min_samples_split / min_sample_leaf / class_wieght
