2. 병합된 데이터를 기준으로 분석합니다. 가장 많이 유통된 상위 3개 상품 소분류에 대해, 상품 판매 금액의 유의미한 차이가 있는지 확인하시오.

```python
pv1 = df1.pivot_table(index='소분류' , values='판매금액', aggfunc='sum').sort_values(by='판매금액', ascending=False).head(3)
pv1
```

```python
top_list = df1['소분류'].value_counts().head(3).index
top_list
```

```python
cond1 = df1['소분류'] == top_list[0]
cond2 = df1['소분류'] == top_list[1]
cond3 = df1['소분류'] == top_list[2]


# 세 집단에 대한 연속형 자료의 대표값 비교(평균/중앙값)
# 2-1 각 세 집단의 연속형 자료에 대한 정규성 검정
# 정규분포 따르는지. -> pvalue < 0.5 => 귀무가설 기각 => 정규분포XX
print(stats.normaltest(df1.loc[cond1]['판매금액']))
print(stats.normaltest(df1.loc[cond2]['판매금액']))
print(stats.normaltest(df1.loc[cond3]['판매금액']))
```

```python
df2 = df1.loc[cond1 | cond2 | cond3]
```

```python
df2.shape
```

```python
groups = df2.groupby('소분류')['판매금액'].apply(list)
stats.kruskal(*groups)
#  pvalue < 0.5 => 귀무가설 기각 => 대표값의 유의미한 차이가 있다.
```

```python
# 2-2 (정규성을 따르지 않는) 세 집단의 연속형 자로ㅛ에 대한 대표값(중앙값)의 유의미한 차이가 있는가?
# 귀무가설 : 모든 집단의 중앙값 같다 / 대립가설 : 모든 집단의 중앙값은 같지 않다(중앙값의 유의미한 차이가 있다)
stats.kruskal(df1.loc[cond1]['판매금액'], df1.loc[cond2]['판매금액'], df1.loc[cond3]['판매금액'])
#  pvalue < 0.5 => 귀무가설 기각 => 대표값의 유의미한 차이가 있다.
```

3. 가장 많이 유통되고 있는 제품에 대해 유통허브에 따라 유통수량의 유의미한 차이가 있는지 확인하시오.

```python
p2 = df1.pivot_table(index='소분류', values='수량', aggfunc='sum').sort_values(by='수량', ascending=False)
p2
```

```python
# 3-1 가장 많이 유통되고 있는 제품을 확인
df1['제품명'].value_counts().head(1)
```

```python
df3 = df1.loc[df1['제품명'] == '2080LINE휴대용세트(브라운)']
df3['유통Hub'].unique()
```

```python
df3.pivot_table(index='유통Hub', values='수량', aggfunc=['sum','mean'])
```

```python
# 3-2 네 집단의 유통 수량의 분포가 정규성을 따르는가
cond_list = ['경상', '서울/수도권', '충청/전라', '강원']
for i in cond_list :
    print(stats.normaltest(df3.loc[df3['유통Hub'] == i]['수량']))
```

```python
groups = df3.groupby('유통Hub')['수량'].apply(list)
stats.kruskal(*groups)
#  pvalue < 0.5 => 귀무가설 기각 => 대표값의 유의미한 차이가 있다.
```

```python
# 3-3 각 지역별 유통 수량의 중앙값 차이는 있다고 볼 수 있는가?
# 귀무가설 : 네 집단 중앙값 모두 같다 / 대립 가설 : 네 집단의 중앙값은 모두 같다고 보기 어렵다 (한 집단이라도 차이가 있다)
stats.kruskal(df3.loc[df3['유통Hub'] == '경상']['수량'], df3.loc[df3['유통Hub'] == '서울/수도권']['수량'],
              df3.loc[df3['유통Hub'] == '충청/전라']['수량'], df3.loc[df3['유통Hub'] == '강원']['수량'])
```

```python
pd.crosstab(df1['채널'], df1['유통Hub'], margins=True)
```

```python
c1 = pd.crosstab(df1['채널'], df1['유통Hub'])
```

```python
# 유통 허브에 따른 유통 채널의비율의 차이가 있는지 검정 (Chi2 Test)
# 귀무가설 : 유통 허브와 유통 채널은 서로 독립이다. (채널 비율 같다) / 대립가설 : 독립이 아니다 (채널 비율이 달리진다)
stats.chi2_contingency(c1)[1] # P.value > 0.05 / 귀무가설 참 => 유통 허브와 유통 채널은 서로 독립이다.
```

6. 데이터가 수집된 기간 내, 유통량이 온라인 소비자들의 수요와 관련이 있는지 확인하고자 한다. 데이터에 가장 많이 유통된 제품의 소분류를 확인하여, 아래 사이트를 활용해 해당 소분류를 소비자들이 검색했을 때, 검색량과 유통수량의 상관성이 있는지 분석하시오.

    - 성별 / 연령대 / 범위에 따라 각각 계산하시오.

```python
import os
```

```python
data_list = os.listdir(r'C:\Users\UserK\Desktop\Ranee\data\네이버데이터랩\네이버데이터랩/') # 특정 폴더 내 파일 명을 확인
```

```python
df_search = pd.DataFrame()
for i in data_list : 
    print(i.split('_')) # [datalab,남성,10대.xlsx]
    dfn = pd.read_excel(r'C:\Users\UserK\Desktop\Ranee\data\네이버데이터랩\네이버데이터랩/' + i, skiprows=6)                        
    dfn['성별'] = i.split('_')[1] # 남성
    dfn['연령'] = i.split('_')[2][:-5] # 40대.xlsx
    df_search = pd.concat([df_search, dfn])
```

```python
df2 = df_search.rename(columns={'바디케어세트' : '검색량'})
```

```python
df2['검색일자'] = pd.to_datetime(df2['날짜'])
```

```python
cond1 = (df1['소분류'] == '바디케어세트')
p2 = df1.loc[cond1].pivot_table(index='공급일자_dt' , values='수량', aggfunc='sum').reset_index()
```

```python
df3 = pd.merge(p2,df2,left_on = '공급일자_dt', right_on ='검색일자' , how='inner')
df3.head()
```

```python
p3 = df3.pivot_table(index='검색일자', values='검색량', aggfunc='sum').reset_index()
m1 = pd.merge(p2,p3, left_on='공급일자_dt', right_on = '검색일자', how='inner')
```

```python
# 상관분석(상관계수 / 모든X가 정규성 -> pearson / 정규성X -> spearman)
print(stats.normaltest(m1['검색량']))
print(stats.normaltest(m1['수량']))
m1[['검색량','수량']].corr(method='spearman')
```

```python
# 귀무가설 : 두 연속형 자료는 상관성이 없다 / 대립가설 : 상관성 있다.
stats.spearmanr(m1['검색량'],m1['수량'])
# pvalue > 0.05 => 귀무가설 참 => 상관성X
```

```python
m1['검색량D5'] = m1['검색량'].shift(5) # 검색한날과 실제 수량은 날짜 차이가 있을것이다 (소비자들 니즈가 있고 (검색량) 5일 후 주문이 높아지더라구요~)
```

```python
m2 = m1.dropna() # 결측값제거
stats.spearmanr(m2['검색량D5'] , m2['수량']) # pvalue < 0.05 => 귀무가설 기각 => 상관성O
```
