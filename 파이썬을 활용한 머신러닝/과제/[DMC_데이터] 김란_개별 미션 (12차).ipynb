### 2. 기업의 설립일을 이용해, 기업의 업력 (설립 후 현재까지 지속된 기간 / 연도)을 계산하고, 대출 승인여부에 따른 기업 업력의 대표값의 차이가 있는지 검정하시오.

```python
df1['설립일_dt'] = pd.to_datetime(df1['설립일'])
```

```python
df1['업력'] = 2024 - df1['설립일_dt'].dt.year
```

```python
df1['업력']
```

```python
cond1 = (df1['진행상태'] == '여신승인')
cond2 = (df1['진행상태'] == '여신거부')
df_pass = df1.loc[cond1]
df_fail = df1.loc[cond2]

# Pvalue(귀무가설이 참이 확률) < 0.05 => 귀무가설(정규분포따름) 기각 => 대립가설 참 => 정규분포 따르지않음
print(stats.normaltest(df_pass['업력']))
print(stats.normaltest(df_fail['업력']))
```

```python
# 귀무가설 : 독립적이다 (중앙값차이가없다)
# 대립가설 : 독립적이지않다 (중앙값차이가 있다)
stats.ranksums(df_pass['업력'],df_fail['업력']) # pvale < 0.05 => 귀무가설 기각 => 대립가설 참 => 대표값 차이가 있다
```

### 3. 기업 평가등급과 대출 승인여부의 유의미한 차이가 있는지 가설 검정을 수행하고, 대출 승인 비율을 각 평가 등급 별로 계산하시오.

```python
df1.info()
```

```python
c1 = pd.crosstab(df1['평가등급'],df1['진행상태'])
stats.chi2_contingency(c1)
# < 0.05 => 귀무가설 기각 => 대립가설 참 => 상관있다.
```

```python
c2 = pd.crosstab(df1['평가등급'],df1['진행상태'], margins=True).reset_index()
c2['여신거부 비율'] = c2['여신거부'] / c2['All'] * 100
c2['여신승인 비율'] = c2['여신승인'] / c2['All'] * 100
c2['투자연계 비율'] = c2['투자연계'] / c2['All'] * 100
c2
```
