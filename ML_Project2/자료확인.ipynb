# 체험팩을 구매한 고객들에 대한 현황 조사

```python
df1['count'] =1
```

```python
df2 = df1.loc[df1['상품명'].str.contains('체험팩')]
```

```python
# 체험팩을 구매한 고객들의 체험팩 구매횟수
pv1 = df2.pivot_table(index='고객번호', values='count' , aggfunc='sum').reset_index()
```

```python
pv1
```

```python
df1.loc[df1['상품명'].str.contains('체험팩')]['고객번호'].tolist()
```

```python
member_number_yam = df1.loc[df1['상품명'].str.contains('체험팩')]['고객번호'].tolist()
cond1 = df1['고객번호'].isin(member_number_yam)
df3 = df1.loc[cond1]
```

```python
df3
```

```python
# 체험팩을 구매한 고객들의 전체 구매횟수
pv2 = df3.pivot_table(index='고객번호', values='count' , aggfunc='sum').reset_index().rename(columns={'count' : 'count_전체'})
```

```python
pv2
```

```python
df4 = pd.merge(pv1,pv2,on='고객번호')
```

```python
df4['차이'] = df4['count_전체'] - df4['count']
df4
```
