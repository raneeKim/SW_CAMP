```python
def make_index(ex, index_name):
    # Index 존재 여부 확인 / business_plan 생성
    if es.indices.exists(index=index_name):
        print('business_plan already exists.')
    else:
        # JSON 파일에서 매핑 불러오기
        with open('mapping.json', 'r') as file:
            body = json.load(file)

        ex.indices.create(index=index_name, body=body)
        print(f'{index_name} creation complete with mapping.')

# Index 생성
make_index(es, 'business_plan')
```

```python
# doc별로 ElasticSearch에 데이터 삽입
def insert_data(es, index_name, data):
    for i, doc in enumerate(data):
        response = es.index(index=index_name, id=f"doc_{i+1}", document=doc)
        print(f"{index_name}에 doc_{i+1} 데이터 삽입 완료:", response)


# data 호출 함수
def get_data(es, index_name, doc_id):
    try:
        response = es.get(index=index_name, id=doc_id)
        return response['_source']
    except Exception as e:
        if e.info.get('status') == 404:
            print(f"{index_name}에서 문서를 찾을 수 없음")
        else:
            print(f"{index_name}에서 오류 발생: {e}")
        return None


# data 삭제 함수
def delete_document(es, index_name, doc_id):
    try:
        response = es.delete(index=index_name, id=doc_id)
        print(f"{index_name}에서 문서 ID '{doc_id}' 삭제 완료:", response)
    except Exception as e:
        print(f"{index_name}에서 문서 ID '{doc_id}' 삭제 오류:", e)
```
